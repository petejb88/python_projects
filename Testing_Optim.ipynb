```
import torch
import torch.optim as optim
```

```
lr = 0.01
a = torch.tensor(4.).requires_grad_(True)
optimizer = optim.Adam([a], lr=lr)
values = [a.item()]
delta_a = []
step = 0
while a > 0.1:
  step += 1
  values.append(a.item())  
  
  loss = a**2
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  delta_a.append((values[-1]-a).item())
  # print("Diff: {}, a = {}".format(round((b-a).item(),3),a))
plt.subplot(1,2,1)
plt.plot(values)
plt.subplot(1,2,2)
plt.plot(delta_a)
print(step)
```
